1. Load `olist_products_dataset.csv`. Fix the two column name typos. Fill null `product_category_name` with `'unknown'`. Confirm 0 nulls remain.

```python
# Load `olist_products_dataset.csv`.
import pandas as pd
from google.colab import drive
import os

#drive.mount('/content/drive')
olist_path = '/content/drive/MyDrive/olist-data'

products    = pd.read_csv(os.path.join(olist_path, '/content/olist_products_dataset.csv'))
print(f"✓ {len(products):,} products loaded")
print()
print(products.shape)
```

```python
# Fix the two column name typos.
products = products.rename(columns={
    'product_name_lenght': 'product_name_length',
    'product_description_lenght': 'product_description_length'
})

print(products.columns.tolist())
```


2. Load `olist_orders_dataset.csv`. Convert all 5 date columns. Calculate `delivery_days`. Answer:
   - What is the median delivery time? *(Expected: 10.0 days)*
   - What % of delivered orders took more than 30 days?
   - What is the longest delivery in days? *(Expected: 209 days)*

```python
# Load `olist_orders_dataset.csv`.
import pandas as pd
from google.colab import drive
import os

#drive.mount('/content/drive')
olist_path = '/content/drive/MyDrive/olist-data'

orders    = pd.read_csv(os.path.join(olist_path, '/content/olist_orders_dataset.csv'))
print(f"✓ {len(orders):,} orders loaded")
print()
print(orders.shape)
```

```python
# Convert all 5 date columns. Calculate delivery_days
delivery_days = pd.to_datetime(orders['order_delivered_customer_date']) - pd.to_datetime(orders['order_purchase_timestamp'])
orders['delivery_days'] = delivery_days.dt.days
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_approved_at'] = pd.to_datetime(orders['order_approved_at'])
orders['order_delivered_carrier_date'] = pd.to_datetime(orders['order_delivered_carrier_date'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])

print("Median delivery time:", orders['delivery_days'].median())
print()
print(f"% of delivered orders took more than 30 days: {(orders['delivery_days'] > 30).mean() * 100:.2f}%")
print()
print("Longest delivery in days:", orders['delivery_days'].max())
```


3. Load `olist_customers_dataset.csv`. Clean `customer_city` (strip + title case). Find the top 10 cities by customer count. *(Expected: Sao Paulo = 15,540)*

```python
# Load `olist_customers_dataset.csv`.
import pandas as pd
from google.colab import drive
import os

#drive.mount('/content/drive')
olist_path = '/content/drive/MyDrive/olist-data'

customers    = pd.read_csv(os.path.join(olist_path, '/content/olist_customers_dataset.csv'))
print(f"✓ {len(customers):,} customers loaded")
print()
print(customers.shape)
```

```python
# Clean `customer_city` (strip + title case)
customers['customer_city'] = customers['customer_city'].str.strip().str.title()

# Find the top 10 cities by customer count
top_10_cities = customers['customer_city'].value_counts().head(10)
print("Top 10 cities by customer count:")
print(top_10_cities)
                                                      # (Expected: Sao Paulo = 15,540)*
```


4. Load `olist_order_reviews_dataset.csv`. Add `has_comment` column. What % of reviews include a comment message? For those that do, what is the average comment length in characters?

```python
# Load `olist_reviews_dataset.csv`.
import pandas as pd
from google.colab import drive
import os

#drive.mount('/content/drive')
olist_path = '/content/drive/MyDrive/olist-data'

reviews    = pd.read_csv(os.path.join(olist_path, '/content/olist_order_reviews_dataset.csv'))
print(f"✓ {len(reviews):,} reviews loaded")
print()
print(reviews.shape)
```

```python
# Add has_comment column
reviews["has_comment"] = reviews["review_comment_message"].notna()

# Percentage of reviews with a comment
pct_with_comment = reviews["has_comment"].mean() * 100

# Average comment length (characters) for reviews with comments
avg_comment_length = (
    reviews.loc[reviews["has_comment"], "review_comment_message"]
    .str.len()
    .mean()
)

print(f"Percentage of reviews with comments: {pct_with_comment:.2f}%")
print()
print(f"Average comment length: {avg_comment_length:.2f} characters")
```
