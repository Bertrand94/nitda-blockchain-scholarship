## Question 1 — Clean Customer City Names

The `customers` DataFrame has a `customer_city` column stored in lowercase with possible extra whitespace.

1. Add a new column `customer_city_clean` = `customer_city` with `.str.strip()` then `.str.title()`
2. Print the first 5 values of `customer_city_clean`

**Expected:** `['Franca', 'Sao Bernardo Do Campo', 'Sao Paulo', 'Mogi Das Cruzes', 'Campinas']`

```python
print(customers['customer_city'].head(5).tolist())
```

```python
# Q1 — Clean customer_city: strip + title case
customers['customer_city_clean'] = customers['customer_city'].str.strip().str.title()

print(customers['customer_city_clean'].head(5).tolist())  # expected: ['Franca', 'Sao Bernardo Do Campo', ...]
```

## Question 2 — Count Cities Containing 'Paulo'

Using the **original** (uncleaned) `customer_city` column, count how many customer rows contain the word `'paulo'` (case-insensitive).

**Expected:** `15,606`

```python
# Q2 — Count rows where customer_city contains 'paulo' (case-insensitive)
paulo_count = customers[customers["customer_city"].str.contains("paulo", case=False, na=False)].shape[0]
print(f"Customers in 'paulo' cities: {paulo_count:,}")  # expected: 15,606
```

## Question 3 — Category Display Names

The `translations` DataFrame has a column `product_category_name_english` with values like `'health_beauty'`.

Add a column `category_display` by:
1. Replacing underscores with spaces: `.str.replace('_', ' ', regex=False)`
2. Applying title case: `.str.title()`

Print the first 5 display names.

**Expected first entries:** `['Health Beauty', 'Computers Accessories', 'Auto', ...]`

```python
translations.head()
```

```python
# Q3 — Create category_display column
translations['category_display'] = translations['product_category_name_english'].str.replace('_', ' ', regex=False).str.title()

print(translations['category_display'].head(5).tolist())
print(f"Total categories: {len(translations)}")  # expected: 71
```
