## Question 2 — Fix the Column Name Typos

The products dataset has two column names with a typo: `product_name_lenght` and `product_description_lenght` (should be `length`).

## Question 3 — Fill Null Categories

## Question 5 — Convert Date Columns

The `orders` DataFrame has 5 date columns stored as strings. Convert all of them to datetime using `pd.to_datetime()` with `errors='coerce'`.

The 5 columns are:
- `order_purchase_timestamp`
- `order_approved_at`
- `order_delivered_carrier_date`
- `order_delivered_customer_date`
- `order_estimated_delivery_date`

After conversion, print the dtype of `order_purchase_timestamp`.  
**Expected:** `datetime64[ns]`

```python
# Q5 — Convert all 5 date columns to datetime
date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in date_cols:
    orders[col] = pd.to_datetime(orders[col], errors='coerce')

print(orders['order_purchase_timestamp'].dtype)  # expected: datetime64[ns]
```

## Question 6 — Calculate Delivery Days

Using the orders DataFrame with datetime columns (from Q5):

1. Add a new column `delivery_days` = `order_delivered_customer_date` minus `order_purchase_timestamp`, converted to integer days using `.dt.days`
2. Filter to delivered orders only: `orders[orders['order_status'] == 'delivered']`
3. Print `.describe().round(1)` on `delivery_days`

**Expected key values:**
- count: `96470.0`
- mean: `12.1`
- median (50%): `10.0`
- max: `209.0`

```python
# Q6 — Calculate delivery_days and get stats for delivered orders
# Your code here — add delivery_days column
orders['delivery_days'] = (orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']).dt.days

# Filter to delivered orders
delivered = orders[orders['order_status'] == 'delivered']

# Your code here — print describe on delivery_days
print(delivered['delivery_days'].describe().round(1))
```

**Task 3:** Load orders. Convert all 5 date columns using pd.to_datetime(errors='coerce'). Calculate delivery_days. Verify: mean ≈ 12.1 days.

```python
import pandas as pd
from google.colab import drive
import os

# drive.mount('/content/drive')
olist_path = '/content/drive/MyDrive/olist-data'

orders_raw = pd.read_csv(os.path.join(olist_path, '/content/olist_orders_dataset.csv'))
```

```python
date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

# Convert date columns to datetime
for col in date_cols:
    orders_raw[col] = pd.to_datetime(orders_raw[col], errors='coerce')

orders_raw['delivery_days'] = (orders_raw['order_delivered_customer_date'] - orders_raw['order_purchase_timestamp']).dt.days

# Filter to delivered orders
delivered = orders_raw[orders_raw['order_status'] == 'delivered']

# Your code here — print describe on delivery_days
print(delivered['delivery_days'].describe().round(1))                      #Verify: mean ≈ 12.1 days.
```

**Task 4 (stretch):** What % of delivered orders took more than 30 days?  
*Hint: `delivered['delivery_days'].gt(30).mean() * 100`*

```python
# (stretch): % orders > 30 days
percentage_over_30 = delivered['delivery_days'].gt(30).mean() * 100
print(f"{percentage_over_30:.2f}%")
```
