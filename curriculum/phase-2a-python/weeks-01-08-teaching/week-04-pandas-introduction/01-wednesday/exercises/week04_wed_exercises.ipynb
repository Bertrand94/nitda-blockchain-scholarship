## Question 2 — Find Missing Values

Use `.isnull().sum()` to count missing values in every column.

**Then answer:** How many orders have **no delivery date** (`order_delivered_customer_date` is null)?

Expected answer: **2,965**

````python
# Q2a: Print null counts for all columns
# .isnull().sum() — exact null count for every column
print(orders.isnull().sum())
print()

# Q2b: Print just the null count for order_delivered_customer_date
print(orders['order_delivered_customer_date'].isnull().sum())
# expected: 2,965
````

## Question 3 — Order Status Breakdown

Use `.value_counts()` to count how many orders have each `order_status`.

Expected output (top values):
```
delivered      96478

```

**Then answer:** How many unique statuses are there? Use `.nunique()`.

## Question 4 — Filter: Canceled or Unavailable Orders

Filter the `orders` DataFrame to only rows where `order_status` is either `'canceled'` **or** `'unavailable'`.

Print the number of matching rows.

Expected: **1,234** (625 canceled + 609 unavailable)

> Hint: Use `.isin()` with a list, or use `|` with two conditions (remember to wrap each in parentheses).

````python
# Q4: Filter to canceled or unavailable orders and count rows
# Filter for canceled orders
canceled = orders[orders['order_status'] == 'canceled']


# Filter for unavailable orders
unavailable = orders[orders['order_status'] == 'unavailable']

# Print the number of matching rows
print(len(canceled) + len(unavailable))
````

## Question 5 — Delivery Rate

Calculate the **percentage of all orders** that have `order_status == 'delivered'`.

Expected: **97.01%**

> Hint: Divide the count of delivered orders by the total number of orders, then multiply by 100.

````python
delivered_orders_count = orders[orders['order_status'] == 'delivered'].shape[0]
total_orders_count = orders.shape[0]
delivered_order = (delivered_orders_count / total_orders_count) * 100

print(f"{delivered_order:.2f}%")
# Expected: 97.01%
````

## Question 6 — Filter to Top 3 Statuses

Filter the `orders` DataFrame to only include rows where `order_status` is one of the **three most common statuses**: `'delivered'`, `'shipped'`, or `'invoiced'`.

Print the shape of the filtered DataFrame.

Expected shape: **(97899, 8)** — that is 96,478 + 1,107 + 314 rows.

> Hint: Use `.isin()` with a list of three status strings.

````python
# Q6: Filter to top 3 statuses and print the shape
top_3_statuses = ['delivered', 'shipped', 'invoiced']
filtered_orders = orders[orders['order_status'].isin(top_3_statuses)]
print(filtered_orders.shape)
````

## Question 7 — Healthy Orders Only

The operations team defines a "healthy" order as one that is **not** `'canceled'` and **not** `'unavailable'`.

Filter `orders` to only healthy orders using the `~` (NOT) operator with `.isin()`.

Print the count of healthy orders.

Expected: **98,207** (99,441 − 625 − 609)

````python
# Q7: Filter to healthy orders using ~ with .isin() and print the count
# given
problem_statuses = ['canceled', 'unavailable']
healthy_orders = orders[~orders['order_status'].isin(problem_statuses)]
print(len(healthy_orders))
````
