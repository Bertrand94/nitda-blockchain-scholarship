**Exercise 9: Safe Float Conversion**

Define a function `safe_float(value, default=0.0)` that tries to convert `value` to a float. If it succeeds, return the float; if it raises a `ValueError`, return `default`.
Test it with `"58.90"` and `"N/A"`.

Expected output:
```
58.9
0.0
```

````python
# Exercise 9
def safe_float(value, default=0.0):
    try:
        return float(value)
    except ValueError:
        return default

# Tests
print(safe_float("58.90"))

print(safe_float("N/A"))
````

**Exercise 10: Safe Integer Conversion**

Write a function `safe_int(value, default=0)` that converts a value to `int` and returns `default` if it fails.
Test it with the following cases:
```python
print(safe_int("42"))      # Expected: 42
print(safe_int("hello"))   # Expected: 0 (default)
print(safe_int("3.5"))     # Expected: 0 (default, because "3.5" is not a valid int)
```

Expected output:
```
42
0
0
```

````python
# Exercise 10
def safe_int(value, default=0):
    try:
        return int(value)
    except ValueError:
        return default

# Tests
print(safe_int("42"))

print(safe_int("hello"))

print(safe_int("3.5"))
````
