9. **Challenge:** Write a function `format_order(order_id, price, freight)` that returns a string. The string should show the first 8 characters of `order_id` followed by `"..."`, then the total cost (`price + freight`) formatted to two decimals, and the freight value. Example: `"Order 4244733e... costs R$72.19 including R$13.29 freight"`. Test your function with the given `product_id`, `price`, and `freight` and print the result.

```python
def format_order(order_id, price, freight):
    total = price + freight
    return f"Order {order_id[:8]}... costs R${total:.2f} including R${freight:.2f} freight"

order_id = "4244733e96bd4903a926d18ff"
price = 58.90
freight = 13.29

result = format_order(order_id, price, freight)
print(result)
```
